```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def triangular(x, a, b, c):
    return np.maximum(0, np.minimum((x - a) / (b - a), (c - x) / (c - b)))

# Definição do eixo X
x = np.linspace(-3500, 3500, 1000)

# Criando uma função de pertinência triangular
y = triangular(x, -1000, 0, 1000)
```

```python
# Plotando o gráfico
plt.plot(x, y, label="Triangular (20, 50, 80)")
plt.xlabel("Variável")
plt.ylabel("Grau de Pertinência")
plt.title("Função de Pertinência Triangular")
plt.legend()
plt.grid()
plt.show()
```
