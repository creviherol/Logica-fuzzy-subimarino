```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def triangular(x, a, b, c):
    return np.maximum(0, np.minimum((x - a) / (b - a), (c - x) / (c - b)))

def triangular2(x1, b, c):
    return (c - x1) / (c - b)

def triangular3(x2, a, b):
    return (x2 - a) / (b - a)

# Definição do eixo X
x = np.linspace(-4000, 4000, 1000)
x1 = np.linspace(-3500, -2000, 1000)
x2 =  np.linspace(2000, 3500, 1000)
x3 =  [-4000, -3500]
x4 =  [3500, 4000]
```

```python
# Plotando o gráfico
# Criando uma função de pertinência triangular
y = [1, 1]
x = [-5, -4]
plt.plot(x, y,color=(0.8, 0.1, 0.1))

x =np.linspace(-4, -3, 1000)
y=triangular2(x,-4,-3)
plt.plot(x, y,color=(0.8, 0.1, 0.1), label="Grande-")

x =np.linspace(-3.5, -2, 1000)
y=triangular(x,-3.5,-2.75,-2)
plt.plot(x, y, color=(1.0, 0.6, 0.2), label="Medio-")

x =np.linspace(-2.5, -1, 1000)
y=triangular(x,-2.5,-1.75,-1)
plt.plot(x, y, color=(1.0, 0.85, 0.23), label="Pequeno-")

x =np.linspace(-1.5, 0, 1000)
y=triangular3(x,-1.5,0)
plt.plot(x, y,color=(0.2, 0.8, 0.2), label="Ajuste Fino-")

x =np.linspace(0, 1.5, 1000)
y=triangular2(x,0,1.5)
plt.plot(x, y,'g', label="Ajuste Fino+")

x =np.linspace(1,2.5, 1000)
y=triangular(x,1,1.75,2.5)
plt.plot(x, y, color=(1.0, 0.92, 0.23), label="Pequeno+")

x =np.linspace(2, 3.5, 1000)
y=triangular(x,2,2.75,3.5)
plt.plot(x, y, color=(1.0, 0.5, 0.0), label="Medio+")

x =np.linspace(3, 4, 1000)
y=triangular3(x,3,4)
plt.plot(x, y, color="r", label="Grande+")

y = [1, 1]
x = [4, 5]
plt.plot(x, y,"r")

plt.xlabel("Diferença")
plt.ylabel("Grau de Pertinência")
plt.title("Potencia da Turbina")
plt.legend(loc="lower right")
plt.grid()
plt.rcParams["figure.figsize"] = (15, 6)
plt.figure()
plt.show()
```
